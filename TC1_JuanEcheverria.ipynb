```python
import pandas as pd

# Cargar ratings.csv
ratings = pd.read_csv("data/ml-latest-small/ratings.csv")

ratings.head()

```

```python
# Promedio de rating y número de reseñas por movieId
top_movies = (ratings.groupby("movieId")["rating"]
              .agg(avg_rating="mean", n="count")
              .query("n >= 50")           # umbral mínimo
              .sort_values("avg_rating", ascending=False)
              .head(5))

top_movies
```
